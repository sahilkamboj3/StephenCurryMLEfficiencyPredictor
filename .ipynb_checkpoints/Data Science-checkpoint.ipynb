```python
df_X = df.loc['fg': 'pts']
df_y = df['plus_minus']
```

```python
ALPHA = 0.05
```

```python
for col in df_X.columns:
    tstat, p = stats.ttest_ind(np.array(df_X[col]), np.array(df_y))
    print(p)
    
    if p < ALPHA:
        print('reject')
    else:
        print('dont reject')
```
