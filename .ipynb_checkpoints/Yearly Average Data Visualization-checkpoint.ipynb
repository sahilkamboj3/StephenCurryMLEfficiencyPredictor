```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
import math
import itertools

style.use('ggplot')
sns.set(palette='dark')
```

```python
df = pd.read_csv('sc_yr_avg.csv')
df.drop(['team_id', 'lg_id', 'pos'], axis=1, inplace=True)
df.drop(df.tail(1).index, inplace=True)
df.head()
```

```python
df['age'].unique()
```

```python
def age_labeler(age):
    if age < 24:
        return 'Young'
    elif age < 28 and age > 23:
        return 'Rising'
    else:
        return 'Prime'

df['age label'] = df['age'].apply(lambda row: age_labeler(row))
```

```python
nums = (df.select_dtypes(exclude='object').columns).tolist()
nums.remove('age')

cats = (df.select_dtypes(include='object').columns).tolist()
```

# Quick General Visualization

```python
rows = 9
columns = 3
width = 20
height = 70
fontsize = 15
```

# Offensive Analysis w/out percents

```python
offensive_cols = nums[nums.index('fg_per_g'):nums.index('orb_per_g') + 1]
offensive_cols.append('pts_per_g')
offensive_cols.remove('fg_pct')
offensive_cols.remove('fg3_pct')
offensive_cols.remove('fg2_pct')
offensive_cols.remove('efg_pct')
offensive_cols.remove('ft_pct')
offensive_cols.remove('orb_per_g')

# These may become repetitive and so removing them makes visualization easier.
offensive_cols.remove('fga_per_g')
offensive_cols.remove('fg3a_per_g')
offensive_cols.remove('fg2a_per_g')
offensive_cols.remove('fta_per_g')

offensive_cols
```

```python
fig = plt.figure(1, figsize=(15, 20))
i = 0

combinations = itertools.combinations(offensive_cols, 2)
combinations = [combo for combo in combinations]

for combo in combinations:
    i += 1
    ax = plt.subplot(math.ceil(len(combinations) / 3), 3, i)
    
    for label in df['age label'].unique():
        data = df[df['age label'] == label]
        sns.regplot(x=combo[0], y=combo[1], data=data, ci=None, label=label)
        
    ax.legend()
```

<h3>Insights</h3>
<p>We see that over time, obviously the number of points Stephen Curry scores per game has been on an incline, we also see that his field goals have also been on the rise. 
    <br>
<li>When he was a young rookie/player, he took just 6-7 field goals per game with just about 2 being 3 point attempts. As he got better at those shots over time, not only did that number increase to around 4 - 5 attempts per game, we also see that it became a bigger part of his game. His number of field goals made has remained fairly constant since his "rising" player years yet we see the number of 2 point field goals he takes slowly reduce 4.5 - 5 in his young to rising years to about 4 - 4.25 per game, which over an 82 game season can add up. 

<br>
<li>We see there is a slight relationship between the the number of free throws made per game  and the number of 2 point field goals made. We can look at the attempted free throws and 2 point field goals to double check this pattern.
<br>
<li>An obvious trend is that over time as Stephen Curry takes more field goals and relies more on his 3 point shot, the more his points per game have increased linearly. If we look at the ft/fg2/fg3 vs pts_per_g graphs, we see the fg3 vs pts_per_g has the strongest, positive relationship which shows that his points per game really increased when he started relying on his 3 point shot more. This means the most important and impactful part of his game is his three point shooting.

```python
sns.regplot(df['fg2a_per_g'], df['fta_per_g'])
```

<li>This graph shows that even if Stephen Curry shoots more 2 point field goals(both mid-range and layups), the number of free throws he shoots doesn't increase which is an ironic relationship since one would assume both would have a positive relationship linearly. This shows that just because Stephen Curry takes more shots doesn't necessarily mean he will go to the free throw line more. It matters more that he makes those shots than how many he takes which may sound obvious but shows he can focus more on his 3 point shooting since that yields his best results, not necessarily an increase in his 2 point field shooting.

```python
sns.regplot(df['pts_per_g'], df['ast_per_g'])
```

<li>This shows that although a relatively weak relationship, we see a trend as Stephen Curry's points per game has increased, he has been able to give out more assists to his teammates although we do see a few outliers above and below the line.

```python
df['ast_tov_ratio'] = df['ast_per_g'] / df['tov_per_g']
```

```python
fig = plt.figure(1, figsize=(15, 5))

ax1 = plt.subplot(1, 2, 1)
sns.regplot(df['ast_per_g'], df['tov_per_g'])

ax2 = plt.subplot(1, 2, 2)
sns.lineplot(df['season'], df['ast_tov_ratio'])
plt.xticks(rotation=75)
```

<li>On the regplot on the left, we see that it may seem that Stephen Curry's assists per game and turnovers per game have a strong, linear relationship but the assist to turnover ratio lineplot states a slightly different case. </li>
    <br>
<li>We see that in his young years, he was not as experienced and therefore, has less assists per turnover per game but he quickly turned that around in his rising years. Increasing it from 1.9 to 2.5, we see a massive improvement in this part of his game. However, starting in 2014, there seems to be a consistent drop in his assist to turnover ratio, meaning he makes less assists per turnover and eventually reaches the same number as his young years. This turn is ironic as this lineplot, ideally, should have a purely positive linear relationship, certainly through his prime years yet it seems to be going in the opposite direction.  

```python
df['fg2_fg_ratio'] = df['fg2_per_g'] / df['fg_per_g']
df['fg3_fg_ratio'] = df['fg3_per_g'] / df['fg_per_g']
```

```python
fig = plt.figure(1, figsize=(15, 14))
i = 0

cols = ['fg2_fg_ratio', 'fg3_fg_ratio', 'fg2_pct', 'fg3_pct']

for col in cols:
    i += 1
    ax = plt.subplot(2, 2, i)
    sns.lineplot(df['season'], df[col])
    plt.xticks(rotation=75)
```

<li>These two graphs are visualizations of how Stephen Curry's game has geared towards his 3 point shot a lot more over time and show the steep incline/decline of his 3 point field goals and 2 point field goals, respectively. We also see despite the number of 3 point shots he's taken, he's always been a consistent 3 point shooter, around 41-45% his whole career. 
<li>Despite reducing the number of 2 point shots he's taken over the years, his effective in his 2-point shots has significantly improved over time, showing that although this isn't a major part of his game, it is still a strong part of it.

```python
fig = plt.figure(1, figsize=(15, 5))
i = 0

cols = ['ft_per_g', 'ft_pct']

for col in cols:
    i += 1
    ax = plt.subplot(1, 2, i)
    sns.lineplot(df['season'], df[col])
    plt.xticks(rotation=75)
```

<p>Despite increasing the number of free throws per game, Stephen Curry's free throw percentage has stayed relatively the same at around 90-92% throughout his career, showing his consistency.

```python
df['ft_pt_ratio'] = df['ft_per_g'] / df['pts_per_g']
```

```python
sns.lineplot(df['season'], df['ft_pt_ratio'])
plt.xticks(rotation=75)
```

<p>This lineplot, paired with the previous free throws per game lineplot, show how over the years, until 2017-2018, free throws has become in increasily larger portion of the points scored by Stephen Curry. Despite the drop in 2018-2019, this consistency shows he continues to find ways to make his way to the free throw line as much as he can.

# Defensive Analysis

```python
defensive_cols = nums[nums.index('drb_per_g'):nums.index('pf_per_g') + 1]
defensive_cols.remove('drb_per_g')
defensive_cols.remove('trb_per_g')
defensive_cols.remove('ast_per_g')
defensive_cols
```

```python
fig = plt.figure(1, figsize=(15, 10))
i = 0

combinations = itertools.combinations(defensive_cols, 2)
combinations = [combo for combo in combinations]

for combo in combinations:
    i += 1
    ax = plt.subplot(math.ceil(len(combinations) / 3), 3, i)
    
    for label in df['age label'].unique():
        data = df[df['age label'] == label]
        sns.regplot(x=combo[0], y=combo[1], data=data, ci=None, label=label)
        
    ax.legend()
```

```python
df['stl_per_f'] = df['stl_per_g'] / df['pf_per_g']
df['blk_per_f'] = df['blk_per_g'] / df['pf_per_g']
```

```python
fig = plt.figure(1, figsize=(15, 5))

ax1 = plt.subplot(1, 2, 1)
sns.lineplot(df['season'], df['stl_per_f'])
plt.xticks(rotation=75)

ax2 = plt.subplot(1, 2, 2)
sns.lineplot(df['season'], df['blk_per_f'])
plt.xticks(rotation=75)
```

<li>We see that up until the 2015-2016 season, Stephen Curry kepts improving on his ability to steal the ball without committing a foul. From 2009-2010 until 2015-2016, he improved by just over 0.5 steals; however, after that season, that statistic has been declining to below his rookie year statistics.
    
<li>On the other hand, Stephen Curry has dramatically improved the number of blocks he gets per foul. Starting off at just 0.06, he has consistently improved it to 0.16 blocks per foul in his last year. This shows over time, he has becoem more efficient at getting blocks despite being one of the shorter players in the league.

```python
df['stops_tov_ratio'] = (df['stl_per_g'] + df['blk_per_g']) / df['tov_per_g']
```

```python
sns.lineplot(df['season'], df['stops_tov_ratio'])
plt.xticks(rotation=75)
```

<li>This graph represents the number of steals and blocks combined versus the number of turnovers, essentially the number of forced turnovers by Stephen Curry vs the number he committed. Although, it seems this ratio is under 1, which makes sense since he is more likely to commit a turnover with the number of minutes he plays and his facilitating role, we see for more the most part, he stays between 0.6 and 0.7 without going too low.

<h5>We see in these defensive statistics that although Stephen Curry isn't one of the taller players in the league, he does have strengths as well as weakness in his defensive game, which eventually balance each other out in the course of the season as he has ups and downs in different parts of his defensive game.

# Rebounding Analysis

```python
rebounding_cols = ['trb_per_g', 'orb_per_g', 'drb_per_g']
```

```python
fig = plt.figure(1, figsize=(15, 5))
i = 0

combinations = itertools.combinations(rebounding_cols, 2)
combinations = [combo for combo in combinations]

for combo in combinations:
    i += 1
    ax = plt.subplot(math.ceil(len(combinations) / 3), 3, i)
    
    for label in df['age label'].unique():
        data = df[df['age label'] == label]
        sns.regplot(x=combo[0], y=combo[1], data=data, ci=None, label=label)
        
    ax.legend()
```

```python
df['drb_ratio'] = df['drb_per_g'] / df['trb_per_g']
df['orb_ratio'] = df['orb_per_g'] / df['trb_per_g']
```

```python
fig = plt.figure(1, figsize=(15, 5))

ax1 = plt.subplot(1, 2, 1)
sns.lineplot(df['season'], df['drb_ratio'], hue=df['age label'])
plt.xticks(rotation=75)

ax2 = plt.subplot(1, 2, 2)
sns.lineplot(df['season'], df['orb_ratio'], hue=df['age label'])
plt.xticks(rotation=75)
```

<ul>This simply shows that although there are ups and downs in the defensive rebound and offensive rebound percents of the total rebounds, Stepehn Curry in his prime years has primarily gotten seen more success in getting more defensive rebounds percentage-wise, as we can see we the near 10% increase in his prime years. This makes sense though since he is a smaller guard, so it is tougher for smaller guards to obtain offensive rebounds so they are more effective on the defensive board.

# Note

Something I have noticed in these graphs is that in several of the graphs, there seems to be a peak around the 2015-2016 year or so and decreases in what I have labeled the prime years. This indicates that either the defenses have been keying in on Stephen Curry so he isn't as effective or he is battling some sort of a lingering injury in his prime years since, with what I have labeled, his prime years would seem to be at the end of what I labeled his rising years.
