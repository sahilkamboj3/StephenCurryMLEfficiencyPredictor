```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
import math
import itertools

style.use('ggplot')
sns.set(palette='dark')
```

```python
df = pd.read_csv('sc_yr_avg.csv')
df.drop(['team_id', 'lg_id', 'pos'], axis=1, inplace=True)
df.drop(df.tail(1).index, inplace=True)
df.head()
```

```python
df['age'].unique()
```

```python
def age_labeler(age):
    if age < 24:
        return 'Young'
    elif age < 28 and age > 23:
        return 'Rising'
    else:
        return 'Prime'

df['age label'] = df['age'].apply(lambda row: age_labeler(row))
```

```python
nums = (df.select_dtypes(exclude='object').columns).tolist()
nums.remove('age')

cats = (df.select_dtypes(include='object').columns).tolist()
```

# General Visualization

```python
rows = 9
columns = 3
width = 20
height = 70
fontsize = 15
```

```python
# Visualize the changes in Stephen Curry's games over time and compare the different parts of his career, taking into the account the y-axis increments.
```

```python
fig = plt.figure(1, figsize=(width, height))
i = 0

for col in nums[:len(nums) // 4]:
    i += 1
    ax = plt.subplot(rows, columns, i)
    sns.lineplot(df['season'], df[col], hue=df['age label'])
    plt.xticks(rotation=75)
    ax.set_xlabel(None)
    ax.set_ylabel(col, fontsize=fontsize)
```

```python
fig = plt.figure(1, figsize=(width, height))
i = 0

for col in nums[len(nums) // 4 : len(nums) // 2]:
    i += 1
    ax = plt.subplot(rows, columns, i)
    sns.lineplot(df['season'], df[col], hue=df['age label'])
    plt.xticks(rotation=75)
    ax.set_xlabel(None)
    ax.set_ylabel(col, fontsize=fontsize)
```

```python
fig = plt.figure(1, figsize=(width, height))
i = 0

for col in nums[len(nums) // 2 : 3 * len(nums) // 4]:
    i += 1
    ax = plt.subplot(rows, columns, i)
    sns.lineplot(df['season'], df[col], hue=df['age label'])
    plt.xticks(rotation=75)
    ax.set_xlabel(None)
    ax.set_ylabel(col, fontsize=fontsize)
```

```python
fig = plt.figure(1, figsize=(width, height))
i = 0

for col in nums[3 * len(nums) // 4 :]:
    i += 1
    ax = plt.subplot(rows, columns, i)
    sns.lineplot(df['season'], df[col], hue=df['age label'])
    plt.xticks(rotation=75)
    ax.set_xlabel(None)
    ax.set_ylabel(col, fontsize=fontsize)
```

```python
# Plotting the correlations of all of the features.

fig = plt.figure(1, figsize=(15, 6))
ax1 = plt.subplot(1, 2, 1)
sns.heatmap(df.corr())

# These shows all the pairs have a correlation of 0.9 or higher.

corr_max = 0.9
ax2 = plt.subplot(1, 2, 2)
sns.heatmap(df.corr() > corr_max)
```

# Offensive Analysis w/out percents

```python
offensive_cols = nums[nums.index('fg_per_g'):nums.index('orb_per_g') + 1]
offensive_cols.append('pts_per_g')
# offensive_cols.append('ast_per_g')
offensive_cols.remove('fg_pct')
offensive_cols.remove('fg3_pct')
offensive_cols.remove('fg2_pct')
offensive_cols.remove('efg_pct')
offensive_cols.remove('ft_pct')
offensive_cols.remove('orb_per_g')

# These may become repetitive.
offensive_cols.remove('fga_per_g')
offensive_cols.remove('fg3a_per_g')
offensive_cols.remove('fg2a_per_g')
offensive_cols.remove('fta_per_g')

offensive_cols
```

```python
fig = plt.figure(1, figsize=(15, 20))
i = 0

combinations = itertools.combinations(offensive_cols, 2)
combinations = [combo for combo in combinations]

for combo in combinations:
    i += 1
    ax = plt.subplot(math.ceil(len(combinations) / 3), 3, i)
    
    for label in df['age label'].unique():
        data = df[df['age label'] == label]
        sns.regplot(x=combo[0], y=combo[1], data=data, ci=None, label=label)
        
    ax.legend()
```
