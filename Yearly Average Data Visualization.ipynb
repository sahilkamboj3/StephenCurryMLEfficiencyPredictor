```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
import math
import itertools

style.use('ggplot')
sns.set(palette='dark')
```

```python
df = pd.read_csv('sc_yr_avg.csv')
df.drop(['team_id', 'lg_id', 'pos'], axis=1, inplace=True)
df.drop(df.tail(1).index, inplace=True)
df.head()
```

```python
df['age'].unique()
```

```python
def age_labeler(age):
    if age < 24:
        return 'Young'
    elif age < 28 and age > 23:
        return 'Rising'
    else:
        return 'Prime'

df['age label'] = df['age'].apply(lambda row: age_labeler(row))
```

```python
nums = (df.select_dtypes(exclude='object').columns).tolist()
nums.remove('age')

cats = (df.select_dtypes(include='object').columns).tolist()
```

# Offensive Analysis w/out percents

```python
offensive_cols = nums[nums.index('fg_per_g'):nums.index('orb_per_g') + 1]
offensive_cols.append('pts_per_g')
offensive_cols.remove('fg_pct')
offensive_cols.remove('fg3_pct')
offensive_cols.remove('fg2_pct')
offensive_cols.remove('efg_pct')
offensive_cols.remove('ft_pct')
offensive_cols.remove('orb_per_g')

# These may become repetitive and so removing them makes visualization easier.
offensive_cols.remove('fga_per_g')
offensive_cols.remove('fg3a_per_g')
offensive_cols.remove('fg2a_per_g')
offensive_cols.remove('fta_per_g')

offensive_cols
```

```python
fig = plt.figure(1, figsize=(15, 20))
i = 0

combinations = itertools.combinations(offensive_cols, 2)
combinations = [combo for combo in combinations]

for combo in combinations:
    i += 1
    ax = plt.subplot(math.ceil(len(combinations) / 3), 3, i)
    
    for label in df['age label'].unique():
        data = df[df['age label'] == label]
        sns.regplot(x=combo[0], y=combo[1], data=data, ci=None, label=label)
        
    ax.legend()
```

<h3>Insights</h3>
<p>We see that over time, obviously the number of points Stephen Curry scores per game has been on an incline, we also see that his field goals have also been on the rise. 
    <br>
<li>When he was a young rookie/player, he took just 6-7 field goals per game with just about 2 being 3 point attempts. As he got better at those shots over time, not only did that number increase to around 4 - 5 attempts per game, we also see that it became a bigger part of his game. His number of field goals made has remained fairly constant since his "rising" player years yet we see the number of 2 point field goals he takes slowly reduces leading to him taking more 3 pointers.

<br>
<li>We see there is a slight relationship between the the number of free throws made per game  and the number of 2 point field goals made. We can look at the attempted free throws and 2 point field goals to further look into this pattern.
<br>
<li>An obvious trend is that over time as Stephen Curry takes more field goals and relies more on his 3 point shot, the more his points per game have increased linearly. If we look at the ft/fg2/fg3 vs pts_per_g graphs, we see the fg3 vs pts_per_g has the strongest, positive relationship which shows that his points per game really increased when he started relying on his 3 point shot more. This means the part of his game that produced the most volume was his 3 point shooting.
<br>
<li>Lastly, we see in the field goals per game and 3 point field goals per game vs free throws per game graph, as Stephen Curry has taken more of those shots per game, his number of free throws per game also increased. This gives a clearer relationship since looking at the free throws and 2 point field goals plot, it didn't really look like it had a relationship and shows indeed, it may be his increased 3 point shooting that has led to more fouls.

```python
fig = plt.figure(1, figsize=(15, 5))

ax = plt.subplot(1, 2, 1)
sns.regplot(df['fg2a_per_g'], df['fta_per_g'])

ax2 = plt.subplot(1, 2, 2)
sns.regplot(df['fg3a_per_g'], df['fta_per_g'])
```

<li>This graph shows that even if Stephen Curry shoots more 2 point field goals(both mid-range and layups), the number of free throws he shoots doesn't increase which is an ironic relationship since one would assume both would have a positive relationship linearly. This shows that just because Stephen Curry takes more shots doesn't necessarily mean he will go to the free throw line more. His 3 point attempts, on the other hand, shows a direct increase with his free throws attempted. This shows he should focus more on his 3 point field goals since those shots clearly correlate with him going to the free throw line more often.

```python
sns.regplot(df['pts_per_g'], df['ast_per_g'])
```

<li>This shows that although a relatively weak relationship, we see a trend as Stephen Curry's points per game has increased, he has been able to give out more assists to his teammates although we do see a few outliers above and below the line. This statistic does rely heavily on the team systems and situations but does show his ability to consistently give out several assists.

```python
df['ast_tov_ratio'] = df['ast_per_g'] / df['tov_per_g']
```

```python
fig = plt.figure(1, figsize=(15, 5))

ax1 = plt.subplot(1, 2, 1)
sns.regplot(df['ast_per_g'], df['tov_per_g'])

ax2 = plt.subplot(1, 2, 2)
sns.lineplot(df['season'], df['ast_tov_ratio'])
plt.xticks(rotation=75)
```

<li>On the regplot on the left, we see that it may seem that Stephen Curry's assists per game and turnovers per game have a strong, linear relationship showing that with just 1 more turnover per game, Curry had increased his assists per game by almost 3 but the assist to turnover ratio lineplot states a slightly different case. </li>
    <br>
<li>We see that in his young years, he was not as experienced and therefore, has less assists per turnover per game but he quickly turned that around in his rising years. Increasing it from 1.9 to 2.5, we see a massive improvement in this part of his game. However, starting in 2014, there seems to be a consistent drop in his assist to turnover ratio, meaning he makes less assists per turnover and eventually reaches the same number as his younger years. This turn is ironic as this lineplot, ideally, should have a purely positive linear relationship, certainly through his prime years yet it seems to be going in the opposite direction. </li>
<br>
<li>This contrast in the two graphs show that perhaps over the years, Stephen Curry actually has become a little efficient in giving out assists to teammates.

```python
df['fg2_fg_ratio'] = df['fg2_per_g'] / df['fg_per_g']
df['fg3_fg_ratio'] = df['fg3_per_g'] / df['fg_per_g']
```

```python
fig = plt.figure(1, figsize=(12, 14.5))
i = 0

cols = ['fg2_fg_ratio', 'fg3_fg_ratio', 'fg2_pct', 'fg3_pct']

for col in cols:
    i += 1
    ax = plt.subplot(2, 2, i)
    sns.lineplot(df['season'], df[col])
    plt.xticks(rotation=75)
```

<br>
<li>We see that over time, the percent of 2 point shots taken by Stephen Curry has been consistently declining and consistently increasing for his 3 point shots. Despite the number of 3 point shots he's taken, he has managed to stay a consistent 3 point shooter, averaging around 41-45% his whole career. 
<li>Despite reducing the number of 2 point shots he's taken over the years, his effective in his 2-point shots has significantly improved over time, showing that although this isn't a major part of his game, it is still a strong part that helps him be effective. 

```python
fig = plt.figure(1, figsize=(15, 5))
i = 0

cols = ['ft_per_g', 'ft_pct']

for col in cols:
    i += 1
    ax = plt.subplot(1, 2, i)
    sns.lineplot(df['season'], df[col])
    plt.xticks(rotation=75)
```

# Defensive Analysis

```python
defensive_cols = nums[nums.index('drb_per_g'):nums.index('pf_per_g') + 1]
defensive_cols.remove('drb_per_g')
defensive_cols.remove('trb_per_g')
defensive_cols.remove('ast_per_g')
defensive_cols
```

```python
fig = plt.figure(1, figsize=(15, 10))
i = 0

combinations = itertools.combinations(defensive_cols, 2)
combinations = [combo for combo in combinations]

for combo in combinations:
    i += 1
    ax = plt.subplot(math.ceil(len(combinations) / 3), 3, i)
    
    for label in df['age label'].unique():
        data = df[df['age label'] == label]
        sns.regplot(x=combo[0], y=combo[1], data=data, ci=None, label=label)
        
    ax.legend()
```

```python
df['stl_per_f'] = df['stl_per_g'] / df['pf_per_g']
df['blk_per_f'] = df['blk_per_g'] / df['pf_per_g']
```

```python
fig = plt.figure(1, figsize=(15, 5))

ax1 = plt.subplot(1, 2, 1)
sns.lineplot(df['season'], df['stl_per_f'])
plt.xticks(rotation=75)

ax2 = plt.subplot(1, 2, 2)
sns.lineplot(df['season'], df['blk_per_f'])
plt.xticks(rotation=75)
```

<li>We see that in the case of Stephen Curry, his steals and blocks tell two different stories. Up until 2015-2016, his kept getting better at getting steals without committing fouls yet after that season, for some reason has declined to his young, early seasons.</li>
    <br>
<li>Blocks, on the other hand, have become a stronger part of his defense. He gradually increased the number of blocks he made per foul by 0.1 blocks per foul over the seasons, which although may not seem like a big jump, is a solid increase for a guard in the league.   

```python
df['stops_tov_ratio'] = (df['stl_per_g'] + df['blk_per_g']) / df['tov_per_g']
```

```python
sns.lineplot(df['season'], df['stops_tov_ratio'])
plt.xticks(rotation=75)
```

<li>This graph represents the number of steals and blocks combined versus the number of turnovers, essentially the number of forced turnovers by Stephen Curry vs the number he committed. Although, it seems this ratio is under 1, which makes sense since he is more likely to commit a turnover with the number of minutes he plays and his facilitating role, we see for more the most part, he consistently stays between 0.6 and 0.7 without going too low, which shows he, although has his stengths and weaknesses, is a consistently willing defender.

# Rebounding Analysis

```python
rebounding_cols = ['trb_per_g', 'orb_per_g', 'drb_per_g']
```

```python
fig = plt.figure(1, figsize=(15, 5))
i = 0

combinations = itertools.combinations(rebounding_cols, 2)
combinations = [combo for combo in combinations]

for combo in combinations:
    i += 1
    ax = plt.subplot(math.ceil(len(combinations) / 3), 3, i)
    
    for label in df['age label'].unique():
        data = df[df['age label'] == label]
        sns.regplot(x=combo[0], y=combo[1], data=data, ci=None, label=label)
        
    ax.legend()
```

<li>We see that defensive rebounds has a closer relationship with total rebounds per gaame which makes sense since for most players, defensive rebounds are the bulk of which they get. 
  <li>We see offensive rebounds doesn't have a relationship with either defensive rebounds or total rebounds since they are either end of the basketball game but we do see Stephen Curry consistently averages between 0.6-0.8 offensive rebounds per game, which is good for a guard in the league.

```python
df['drb_ratio'] = df['drb_per_g'] / df['trb_per_g']
df['orb_ratio'] = df['orb_per_g'] / df['trb_per_g']
```

```python
fig = plt.figure(1, figsize=(15, 5))

ax1 = plt.subplot(1, 2, 1)
sns.lineplot(df['season'], df['drb_ratio'], hue=df['age label'])
plt.xticks(rotation=75)

ax2 = plt.subplot(1, 2, 2)
sns.lineplot(df['season'], df['orb_ratio'], hue=df['age label'])
plt.xticks(rotation=75)
```

<li>These two lineplots show Stephen Curry, especially in his prime years, has primarily gotten seen more success in getting more defensive rebounds percentage-wise, as we can see we the near 10% increase in his prime years. This makes sense though since he is a smaller guard, so it is tougher for smaller guards to obtain offensive rebounds so they are more effective on the defensive board.

<li>Stephen Curry has become less efficient at giving out assists to his teammates over time.

<li>Despite relying less on his 2-point shooting, he has continued to become more efficient at those shots.
<li>Ironically, Stephen Curry has found more success at going to the free throw line with his increased 3-point shooting rather than increased 2-point shooting. Due to this relationship, he has significantly increased the number of times he goes to and makes shots at the free throw line, all while consistently making them. 
<li>We see that offensively, Stephen Curry best year is in <strong>2015-2016</strong>, where we see this season, in multiple graphs, is either the peak in the graph or one of the season where the graph peaks.

<h3>Defense

<li>Stephen Curry, over the years, has become better at blocking shots and worse at getting steals.
<li>He has become more efficient at grabbing defensive rebounds and slightly less at grabbing offensive rebounds.  
<li>Despite this shift in efficiency of grabbing rebounds, Stephen Curry has continued to increase the number of total rebounds and defensive rebounds he gets per game and maintained that high level.
