```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style

style.use('ggplot')
sns.set(palette='terrain')
```

```python
df = pd.read_csv('sc_yr_avg.csv')
df.drop(['team_id', 'lg_id', 'pos'], axis=1, inplace=True)
df.drop(df.tail(1).index, inplace=True)
df.head()
```

```python
df['age'].unique()
```

```python
def age_labeler(age):
    if age < 24:
        return 'Young'
    elif age < 28 and age > 23:
        return 'Rising'
    else:
        return 'Prime'

df['age label'] = df['age'].apply(lambda row: age_labeler(row))
```

```python
nums = (df.select_dtypes(exclude='object').columns).tolist()
nums.remove('age')

cats = (df.select_dtypes(include='object').columns).tolist()
```

```python
rows = 9
columns = 3
width = 20
height = 70
fontsize = 15
```

```python
fig = plt.figure(1, figsize=(width, height))
i = 0

for col in nums[:len(nums) // 4]:
    i += 1
    ax = plt.subplot(rows, columns, i)
    sns.lineplot(df['season'], df[col])
    plt.xticks(rotation=75)
    ax.set_xlabel(None)
    ax.set_ylabel(col, fontsize=fontsize)
```

```python
fig = plt.figure(1, figsize=(width, height))
i = 0

for col in nums[len(nums) // 4 : len(nums) // 2]:
    i += 1
    ax = plt.subplot(rows, columns, i)
    sns.lineplot(df['season'], df[col])
    plt.xticks(rotation=75)
    ax.set_xlabel(None)
    ax.set_ylabel(col, fontsize=fontsize)
```

```python
fig = plt.figure(1, figsize=(width, height))
i = 0

for col in nums[len(nums) // 2 : 3 * len(nums) // 4]:
    i += 1
    ax = plt.subplot(rows, columns, i)
    sns.lineplot(df['season'], df[col])
    plt.xticks(rotation=75)
    ax.set_xlabel(None)
    ax.set_ylabel(col, fontsize=fontsize)
```
